```python
import glob
import os
import shutil

train = glob.glob("nipa_dataset/train/*")

classes = set()

base_dir = "nipa_dataset/train"

for t in train:
    
    cls_name = "_".join(t.split("\\")[-1].split("_")[:2])
    classes.add(cls_name)

print(classes)
print(len(classes))


# directory generate

#for c in classes:
#    os.mkdir(f"{base_dir}/{c}")
    
for t in train:
    
    splitted = t.split("\\")[-1].split("_")
    
    if len(splitted) == 2:
        continue
    
    file_name = splitted[-1]
    print(file_name)
    cls_name = "_".join(t.split("\\")[-1].split("_")[:2])    
    shutil.move(t, f"{base_dir}/{cls_name}/{file_name}")
```
